```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from tensorflow import keras
import pandas as pd
from sklearn.preprocessing import StandardScaler

fashion_test= pd.read_csv(r"C:\Users\admin\Desktop\hw10-2\fashion-mnist_test.csv")
fashion_train= pd.read_csv(r"C:\Users\admin\Desktop\hw10-2\fashion-mnist_train.csv")



X = np.array(fashion_test.loc[:, fashion_test.columns != 'label'])
y = fashion_test["label"]


# X = X.astype("float32")

# X = X.reshape(X.shape[0], 28, 28, 1)

# X = X / 255

plt.figure(figsize=(20,5))

for i in range(4):
    plt.subplot(1,4,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X[i].reshape(28,28), cmap="gray")
    plt.title("Resmin Etiketi: {}".format(y[i]))
plt.show()

std = StandardScaler()
X_std = std.fit_transform(X)

pca = PCA(n_components=2)
pca_components = pca.fit_transform(X_std)



plt.figure(figsize=(8,4))
colours = ["r","b","g","c","m","y","k","r","burlywood","chartreuse"]
for i in range(100):
    plt.text(pca_components[i, 0], pca_components[i, 1], str(y[i]),
             color=colours[int(y[i])],
             fontdict={'weight': 'bold', 'size': 50}
        )

plt.xticks([])
plt.yticks([])
plt.show()

```
